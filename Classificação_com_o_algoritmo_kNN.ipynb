```python
#Divisão dos dados em conjuntos de treinamento e teste.

X_train_2F_Original = df_TrainingData_2F_Original[['total.sulfur.dioxide', 'citric.acid']]
X_test_2F_Original = df_TestingData_2F_Original[['total.sulfur.dioxide', 'citric.acid']]
y_train_2F_Original = df_TrainingData_2F_Original[['class']]
y_test_2F_Original = df_TestingData_2F_Original[['class']]
```

Treine modelos usando o algoritmo KNN para o conjunto de dados Dados_Originais_2Features (não normalizados), variando o valor de k (número de vizinhos mais próximos) entre 1, 3, 5, 7. Para cada modelo treinado, avalie seu desempenho nos dados de teste, reportando a acurácia

```python
# Importar KNeighbors ClaSSifier do sklearn
from sklearn.neighbors import KNeighborsClassifier

for n_neighbors in [1,3,5,7]:
  # inicializando o modelo
  knn = KNeighborsClassifier(n_neighbors, metric='euclidean')

  # Fitting o modelo no set de training
  knn.fit(X_train_2F_Original, y_train_2F_Original)

  # Predição test-set resultados
  y_pred_2F_Original = knn.predict(X_test_2F_Original)

  from sklearn.metrics import accuracy_score
  # Verificação da Acuracia Check accuracy score
  print('Acuracia do Modelo para k={}: {:.4f}'.format(n_neighbors, accuracy_score(y_test_2F_Original, y_pred_2F_Original)))
```

```python
#Divisão dos dados em conjuntos de treinamento e teste.

X_train_11F_Norm = df_TrainingData_11F_Norm[['fixed.acidity',	'volatile.acidity',	'citric.acid',	'residual.sugar',	'chlorides',	'free.sulfur.dioxide',	'total.sulfur.dioxide',	'density',	'pH',	'sulphates',	'alcohol']]
X_test_11F_Norm = df_TestingData_11F_Norm[['fixed.acidity',	'volatile.acidity',	'citric.acid',	'residual.sugar',	'chlorides',	'free.sulfur.dioxide',	'total.sulfur.dioxide',	'density',	'pH',	'sulphates',	'alcohol']]
y_train_11F_Norm = df_TrainingData_11F_Norm[['class']]
y_test_11F_Norm = df_TestingData_11F_Norm[['class']]
```

```python
# Verificar tamanhos dos X_train and X_test

X_train_11F_Norm.shape, X_test_11F_Norm.shape
```

```python
# Verificar tamanhos dos y_train and y_test

y_train_11F_Norm.shape, y_test_11F_Norm.shape
```

```python
n_neighbors =5
# Inicializar o modelo
knn_2F_Norm = KNeighborsClassifier(n_neighbors, metric='euclidean')

# Fitting do modelo
knn_2F_Norm.fit(X_train_2F_Norm, y_train_2F_Norm)

# Predição test-set resultados
y_pred_2F_Norm = knn_2F_Norm.predict(X_test_2F_Norm)

# Verificar o Score de acuracidade do modelo
print('Acuracidade do modelo para k={}: {:.4f} usando os Dados_Normalizados_2Features'.format(n_neighbors, accuracy_score(y_test_2F_Norm, y_pred_2F_Norm)))

# Obtenha os vizinhos mais próximos para o conjunto de teste
distances_2F_Norm, indices_2F_Norm = knn_2F_Norm.kneighbors(X_test_2F_Norm)

# Imprima as distâncias e índices dos vizinhos mais próximos
print(f"\nDistâncias usando os Dados_Normalizados_2Features:\n{distances_2F_Norm} ")
print(f"\nÍndices usando os Dados_Normalizados_2Features:\n{indices_2F_Norm}")

n_neighbors =5
# Inicializar o modelo
knn_11F_Norm = KNeighborsClassifier(n_neighbors, metric='euclidean')

# Fitting do modelo
knn_11F_Norm.fit(X_train_11F_Norm, y_train_11F_Norm)

# Predição test-set resultados
y_pred_11F_Norm = knn_11F_Norm.predict(X_test_11F_Norm)

# Verificar o Score de acuracidade do modelo
print('Acuracidade do modelo para k={}: {:.4f} usando os Dados_Normalizados_11Features'.format(n_neighbors, accuracy_score(y_test_11F_Norm, y_pred_11F_Norm)))

# Obtenha os vizinhos mais próximos para o conjunto de teste
distances_11F_Norm, indices_11F_Norm = knn_11F_Norm.kneighbors(X_test_11F_Norm)

# Imprima as distâncias e índices dos vizinhos mais próximos
print(f"\nDistâncias usando os Dados_Normalizados_11Features:\n{distances_11F_Norm} ")
print(f"\nÍndices usando os Dados_Normalizados_11Features:\n{indices_11F_Norm}")

```
